```python
def Checks_All_List(df):
    lista =list( df['Symptom_1'].unique())
    for i in range(1,8):
        list_ =  list(df[f'Symptom_{i}'].unique())
        for j in list_:
            
            if j not in lista:
                # print(f"for {i} number we are in {j}")
                lista.append(j)
            else:
                continue
    return lista
        
```
